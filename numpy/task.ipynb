```python
values = np.array([2, 4, 6, 8, 10])

sqrt_values = np.sqrt(values)

exp_values = np.exp(values)


log_values = np.log(values)


total_sum = np.sum(values)


cum_sum = np.cumsum(values)

print(f"Original: {values}")
print(f"Square Root: {sqrt_values}")
print(f"Exponential: {exp_values}")
print(f"Natural Log: {log_values}")
print(f"Total Sum: {total_sum}")
print(f"Cumulative Sum: {cum_sum}")
```

```python
import numpy as np

data = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# 1. Row-wise sum
row_sum = np.sum(data, axis=1)

# 2. Column-wise sum
col_sum = np.sum(data, axis=0)

# 3. Minimum value
min_val = np.min(data)

# 4. Maximum value
max_val = np.max(data)

# 5. Overall mean
overall_mean = np.mean(data)

print(f"Row sums: {row_sum}\nCol sums: {col_sum}\nMin: {min_val}\nMax: {max_val}\nMean: {overall_mean}")
```

```python
marks = np.array([78, 85, 90, 66, 72, 88, 95, 60])

# 1-4. Stats
mean_val = np.mean(marks)
median_val = np.median(marks)
variance_val = np.var(marks)
std_dev = np.std(marks)

# 5-6. Min, Max, and Range
min_marks = np.min(marks)
max_marks = np.max(marks)
marks_range = max_marks - min_marks

print(f"Mean: {mean_val}, Median: {median_val}, Std Dev: {std_dev:.2f}")
print(f"Range: {marks_range} ({min_marks} to {max_marks})")
```

```python
# 1. Sort the array
sorted_marks = np.sort(marks)

# 2. Find Percentiles
p25 = np.percentile(marks, 25)
p50 = np.percentile(marks, 50) # Same as Median
p75 = np.percentile(marks, 75)

# 3. Count students above average
avg = np.mean(marks)
above_avg_count = np.sum(marks > avg)

print(f"Sorted: {sorted_marks}")
print(f"25th: {p25}, 50th: {p50}, 75th: {p75}")
print(f"Students above average: {above_avg_count}")
```

```python
sales = np.array([1200,1500,900,2000,1800,1700,1600])
total_weekly_sales = np.sum(sales)
avg_daily_sales = np.mean(sales)
highest_sales = np.max(sales)
lowest_sales=  np.min(sales)
standard = np.std(sales)
more_avg = sales>avg_daily_sales
print(sales)
print(f"Total Weekly Sales: {total_weekly_sales}")
print(f"Average Daily Sales: {avg_daily_sales}")
print(f"Highest Sales: {highest_sales}")
print(f"Lowest Sales: {lowest_sales}")
print(f"Standard Deviation: {standard}")
print(f"Days with above average sales: {np.sum(more_avg)}")
```
