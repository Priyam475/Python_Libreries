```python

marks_list = [78, 85, 90, 66, 72]
marks = pd.Series(marks_list)

# 4. Print Metadata
print("Series values:\n", marks.values)
print("\nIndex:\n", marks.index)
print("\nData type:\n", marks.dtype)

# 5. Accessing Elements
print("\nFirst element:", marks[0])
print("Last two elements:\n", marks.tail(2)) 
# Alternatively, using slicing: marks[-2:]
```

```python
# 1. Add 5 grace marks
add_grace = marks + 5
print("After adding 5 grace marks:\n", add_grace)

# 2. Subtract 2 marks
sub_marks = marks - 2
print("\nAfter subtracting 2 marks:\n", sub_marks)

# 3. Multiply by 1.05 (5% increase)
mult_marks = marks * 1.05
print("\nAfter multiplying by 1.05:\n", mult_marks)

# 4. Divide by 2
div_marks = marks / 2
print("\nAfter dividing by 2:\n", div_marks)
```

```python
# 1. Statistical aggregates
print(f"Max: {marks.max()}, Min: {marks.min()}")
print(f"Sum: {marks.sum()}, Mean: {marks.mean()}")

# 2. Lambda function to check pass/fail
pass_status = marks.apply(lambda x: "Pass" if x >= 70 else "Fail")
print("\nPass Status:\n", pass_status)

# 3. Count how many passed
passed_count = (marks >= 70).sum()
print(f"\nNumber of students who passed: {passed_count}")
```

```python
# Create DataFrame
sales_dict = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}
sales = pd.DataFrame(sales_dict)

# Analysis
total_rev = sales['Revenue'].sum()
avg_daily = sales['Revenue'].mean()
max_rev_day = sales.loc[sales['Revenue'].idxmax(), 'Day']
above_avg_days = sales[sales['Revenue'] > avg_daily]

# Plotting
sales.plot(kind='line', x='Day', y='Revenue', marker='o', title='Revenue vs Day')

print(f"Total Revenue: {total_rev}")
print(f"Highest Revenue Day: {max_rev_day}")
```
